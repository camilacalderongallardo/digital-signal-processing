# Signal Filtering and Spectrum Analysis

This independent demonstration generates a synthetic sensor signal, analyzes its spectrum, designs an FIR low-pass filter, and compares the signal before and after filtering.

```python
import sys
from pathlib import Path

project_root = Path.cwd().parent
sys.path.insert(0, str(project_root))

import matplotlib.pyplot as plt
import numpy as np

from src.dsp_tools import (
    apply_fir_filter,
    design_lowpass_fir,
    frequency_response,
    generate_test_signal,
    single_sided_spectrum,
)
```

```python
sample_rate = 2000
t, clean, measured = generate_test_signal(sample_rate=sample_rate)

plt.figure(figsize=(10, 4))
shown = t <= 0.05
plt.plot(t[shown], measured[shown], label='Measured signal')
plt.plot(t[shown], clean[shown], label='Desired component')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.title('Synthetic sensor signal')
plt.legend()
plt.show()
```

```python
taps = design_lowpass_fir(sample_rate, cutoff_frequency=250, number_of_taps=81)
filtered = apply_fir_filter(measured, taps)

frequency_before, amplitude_before = single_sided_spectrum(measured, sample_rate)
frequency_after, amplitude_after = single_sided_spectrum(filtered, sample_rate)

plt.figure(figsize=(10, 4))
plt.semilogy(frequency_before, np.maximum(amplitude_before, 1e-8), label='Before')
plt.semilogy(frequency_after, np.maximum(amplitude_after, 1e-8), label='After')
plt.xlim(0, 800)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.title('Spectrum before and after filtering')
plt.legend()
plt.show()
```

```python
frequency, response = frequency_response(taps, sample_rate)

plt.figure(figsize=(10, 4))
plt.plot(frequency, 20 * np.log10(np.maximum(np.abs(response), 1e-8)))
plt.axvline(250, linestyle='--')
plt.ylim(-100, 5)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude (dB)')
plt.title('FIR low-pass filter response')
plt.show()
```
